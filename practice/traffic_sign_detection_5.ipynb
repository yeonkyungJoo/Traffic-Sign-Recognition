```python
import cv2
import numpy as np

img_rgb = cv2.imread('./sample3_515.jpg') # 이미지 읽기
# print(image.shape) # (1080, 1920, 3)
img_rgb = cv2.resize(img_rgb, (img_rgb.shape[1]//2, img_rgb.shape[0]//2))
cv2.namedWindow("Original Image",cv2.WINDOW_NORMAL)

img = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)
cv2.imshow("img1", img)

img = cv2.bilateralFilter(img,9,105,105)
cv2.imshow("img2", img)

r,g,b=cv2.split(img)
equalize1= cv2.equalizeHist(r)
equalize2= cv2.equalizeHist(g)
equalize3= cv2.equalizeHist(b)
equalize=cv2.merge((r,g,b))
cv2.imshow("equalize1", equalize)
equalize = cv2.cvtColor(equalize,cv2.COLOR_RGB2GRAY)
cv2.imshow("equalize2", equalize)


ret,thresh_image = cv2.threshold(equalize,0,255,cv2.THRESH_OTSU+cv2.THRESH_BINARY)
cv2.imshow("thresh_image", thresh_image)
equalize= cv2.equalizeHist(thresh_image)
cv2.imshow("equalize3", equalize)


canny_image = cv2.Canny(equalize,250,255)
canny_image = cv2.convertScaleAbs(canny_image)
kernel = np.ones((3,3), np.uint8)
dilated_image = cv2.dilate(canny_image,kernel,iterations=1)

contours, hierarchy = cv2.findContours(dilated_image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours= sorted(contours, key = cv2.contourArea, reverse = True)[:10]
c=contours[0]
print(cv2.contourArea(c))
final = cv2.drawContours(img, [c], -1, (255,0, 0), 3)

mask = np.zeros(img_rgb.shape,np.uint8)
new_image = cv2.drawContours(mask,[c],0,255,-1,)
new_image = cv2.bitwise_and(img_rgb, img_rgb, mask = equalize)

cv2.namedWindow("new",cv2.WINDOW_NORMAL)
cv2.imshow("new",new_image)

cv2.imshow("Original Image",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pylab as plt


# 색상별 영역 지정
red1 = np.array([0, 50, 50])
red2 = np.array([15, 255, 255])
red3 = np.array([165, 50, 50])
red4 = np.array([180, 255, 255])

img = cv2.imread('./sample3_515.jpg')
img = cv2.resize(img, (img.shape[1]//2, img.shape[0]//2), 
                 interpolation = cv2.INTER_AREA)
h, w = img.shape[:2]
# cv2. imshow('img', img)

# HSV 영상으로 변환
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# cv2. imshow('hsv', hsv)

img = cv2.bilateralFilter(img,9,105,105)
r,g,b=cv2.split(img)
equalize1= cv2.equalizeHist(r)
equalize2= cv2.equalizeHist(g)
equalize3= cv2.equalizeHist(b)
# equalize=cv2.merge((r,g,b))
# cv2.imshow('equalize', equalize)
equalize=cv2.merge((r,g,b))

# 색상에 따른 마스크 생성
mask_red1 = cv2.inRange(hsv, red1, red2)
# cv2.imshow('mask_red1', mask_red1)
mask_red2 = cv2.inRange(hsv, red3, red4)
# cv2.imshow('mask_red2', mask_red2)

mask_red = mask_red1 + mask_red2
cv2.imshow('mask_red', mask_red)

kernel = np.ones((11, 11), np.uint8)
mask_red1 = cv2.morphologyEx(mask_red1, cv2.MORPH_OPEN, kernel)
mask_red1 = cv2.morphologyEx(mask_red1, cv2.MORPH_CLOSE, kernel)
cv2.imshow('mask_red1', mask_red)
mask_red2 = cv2.morphologyEx(mask_red2, cv2.MORPH_OPEN, kernel)
mask_red2 = cv2.morphologyEx(mask_red2, cv2.MORPH_CLOSE, kernel)
cv2.imshow('mask_red2', mask_red)

# mask_red = mask_red1 + mask_red2
# cv2.imshow('mask_red', mask_red)

# 색상별 마스크로 색상만 추출
res_red1 = cv2.bitwise_and(img, img, mask = mask_red1)
# cv2.imshow('res_red1', res_red1)
res_red2 = cv2.bitwise_and(img, img, mask = mask_red2)
# cv2.imshow('res_red2', res_red2)
res_red = cv2.bitwise_or(res_red1, res_red2)
cv2.imshow('res_red', res_red)

cv2.waitKey()
cv2.destroyAllWindows()
```
